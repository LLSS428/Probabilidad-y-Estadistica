```python
import random
# Vector que va a contener 100 edades generadas aleatoriamente
edades=[]
for x in range(100):
  #Rango de edades entre 18 y 70 a침os
  rdm=random.randint(18,70)
  edades.append(rdm)
print(edades)
```

```python
# Con el paquete de pandas se hace un data frame con los numeros aleatorios
import pandas as pd
df=pd.DataFrame(edades,columns=['Edad'])
df.head()
```

```python
# Se va a calcular la tabla de disrtibuci칩n de frecuencias
tabla_frecuencias=df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns=['Edad','Frecuencia']
tabla_frecuencias=tabla_frecuencias.sort_values(by='Edad')
print("Tabla de Distribuci칩n de Frecuencias:")
tabla_frecuencias
```

```python
# Frecuencia relativa
tabla_frecuencias['Frec Relativa']=tabla_frecuencias['Frecuencia']/100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual']=tabla_frecuencias['Frecuencia']/100*100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada']=tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

```python
# Libreria qde funciones matematicas
import math
import matplotlib.pyplot as plt
```

```python
print(edades)
```

```python
# Se crea un DataFrame con las edades
df=pd.DataFrame({'Edad':edades})
df
```

```python
# Se calcula el n칰mero de intervalos utilizando la regla de Strages
n=len(edades)
k=int(1+math.log2(n))
print("Valor de n=",n,"\nN칰mero de intervalos=",k)
```

```python
tabla_frecuencia=pd.cut(df['Edad'],bins=k,include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns=['Clase','Frec']
tabla_frecuencia
```

```python
# Se ordena la tabla de frecuencia por el intervalo
tabla_frecuencia=tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
tabla_frecuencia['Frec Rel']=tabla_frecuencia['Frec']/n
tabla_frecuencia['Frec Porc']=tabla_frecuencia['Frec']/n*100
tabla_frecuencia['Frec Ac']=tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Frec Rel Ac']=tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Porc Ac']=tabla_frecuencia['Frec Porc'].cumsum()
print(tabla_frecuencia)
```
