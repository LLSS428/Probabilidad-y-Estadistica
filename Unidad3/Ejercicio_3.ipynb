```python
import matplotlib.pyplot as plt
import numpy as np
import math
# Valores de la variable aleatoria
x = [0, 1, 2, 3, 4, 5]
p = [0.05, 0.10, 0.25, 0.30, 0.20, 0.10]
# Verificar suma = 1
print("Suma de probabilidades:", sum(p))

valor_esperado=0
for i in range(len(x)):
    xi = x[i]
    pi = p[i]
    producto = xi * pi
    valor_esperado += producto
    print(f"x= {xi}, p={pi}, x*p={producto:.2f}")
print (f"\nValor esperado (Esperanza): {valor_esperado:.2f}")

# Calcular E(X)
esperanza=0
for i in range(len(x)):
    esperanza += x[i] * p[i]
# Calcular E(x'2)
esperanza_cuadrado=0
for i in range(len(x)):
    esperanza_cuadrado += (x[i] ** 2) * p[i]
# Calcular la varianza y desviación estandar
varianza = esperanza_cuadrado - (esperanza ** 2)
desviación = math.sqrt(varianza)
# Mostrar resultados
print(f"E(X) = {esperanza:.2f}")
print(f"E(X^2) = {esperanza_cuadrado:.2f}")
print(f"Varianza = {varianza}")
print(f"Desviación estandar = {desviación:.2f}")

# Graficar distribución
plt.stem(x, p)
plt.title("Distribución de Probabilidad")
plt.xlabel("Ud. Vendidas")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
print("\n")

x = np.array([0, 1, 2, 3, 4, 5])
p = np.array([0.05, 0.10, 0.25, 0.30, 0.20, 0.10])

# Cálculo del valor esperado (media)
esperanza = np.sum(x * p)

# Cálculo de la esperanza del cuadrado
esperanza_cuadrado = np.sum(x**2 * p)

# Cálculo de la varianza y desviación estándar
varianza = esperanza_cuadrado - esperanza**2
desviacion = np.sqrt(varianza)

# Mostrar los resultados en consola
print(f"E(X) = {esperanza:.2f}")
print(f"E(X^2) = {esperanza_cuadrado:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviación estándar = {desviacion:.2f}")
print("\n")

# Crear la gráfica tipo 'stem' (tallos y hojas)
plt.stem(x, p, basefmt="", linefmt='b-', markerfmt='bo')
plt.title("Distribución de Probabilidad Discreta\ncon Media y Desviación Estándar")
plt.xlabel("Valores de la variable aleatoria X")
plt.ylabel("P(X)")
plt.grid(True, linestyle='--', alpha=0.6)

# Dibujar línea vertical en la media
plt.axvline(x=esperanza, color='red', linestyle='--', label=f"Media (E[X]) = {esperanza:.2f}")

# Resaltar el rango de ±1 desviación estándar
plt.axvspan(esperanza - desviacion, esperanza + desviacion, color='red', alpha=0.2,
            label=f"±1 Desv. Estándar ($\sigma$ = {desviacion:.2f})")

# Mostrar leyenda y límites
plt.legend()
plt.ylim(0, max(p) + 0.05)
plt.xticks(x)

# Mostrar la gráfica
plt.show()
```
