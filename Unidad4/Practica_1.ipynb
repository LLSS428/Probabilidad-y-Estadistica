```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar datos para la distribución normal
media = 70       # Media
desv_std = 4     # Desviación estándar
num_datos = 500

# Generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media, desv_std, num_datos)
```

```python

# Crear un rango de valores para crear una densidad
# El rango va de la media - 3*desv_std hasta la media + 3*desv_std
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

# Calcular la PDF de la distribución normal (fórmula explícita, aunque 'norm.pdf' lo hace más fácil)
# pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp( - (x - media)**2 / (2 * desv_std**2))
# Usando scipy para el cálculo:
pdf = norm.pdf(x, media, desv_std)
print("Imprimir los primeros 50 Valores de x", x[0:50], ", y los primeros 50 pdf", pdf[0:50])
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, color='b') # Histograma de los datos
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
# Visualizar la PDF
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Normal({media: .1f}; $\sigma$: {ds: .1f})'.format(media=np.round(media, 2), ds=np.round(desv_std, 2)))
plt.title('Función de Densidad de Probabilidad (PDF) - Distribución Normal')
plt.xlabel('Pesos en Kg de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar histograma combinado
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b') # Histograma de los datos, con densidad=True
plt.plot(x, pdf, 'k', linewidth=2) # Curva de la PDF
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```
