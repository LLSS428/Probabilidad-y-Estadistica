```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parámetros de la distribución
media = 70       # Media
desv_std = 4     # Desviación estándar
num_datos = 500  # Número de estudiantes

# Generamos los datos de manera aleatoria
datos = np.random.normal(media, desv_std, num_datos)

# Crear rango de valores para graficar la densidad
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)
pdf = norm.pdf(x, media, desv_std)

# --- Cálculo de probabilidades ---
lim_inferior = 68
lim_superior = 72

# Probabilidad acumulada entre los límites
probabilidad = norm.cdf(lim_superior, media, desv_std) - norm.cdf(lim_inferior, media, desv_std)

# Número esperado de estudiantes
num_estudiantes = probabilidad * num_datos

print(f"Probabilidad de que un estudiante pese entre {lim_inferior} y {lim_superior} kg: {probabilidad:.4f}")
print(f"Número esperado de estudiantes en ese rango (de {num_datos}): {num_estudiantes:.2f}")

# --- Gráficas ---
# Histograma con curva PDF
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b', label='Datos simulados')
plt.plot(x, pdf, 'k', linewidth=2, label='Distribución Normal')

# Resaltar el área entre 68 y 72 kg
x_rango = np.linspace(lim_inferior, lim_superior, 100)
plt.fill_between(x_rango, norm.pdf(x_rango, media, desv_std), color='orange', alpha=0.5, label='68 - 72 kg')

plt.title('Distribución Normal del Peso de Estudiantes')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
