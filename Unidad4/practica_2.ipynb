```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

plt.figure(figsize=(6,4))
# Parámetros de la distribución normal
media = 70
desv_std = 4
total_estudiantes = 500

# VARIABLE para Cálculo de la probabilidad de que un estudiante que pese menos de 75 kg
peso_limite_75 = 75

# Calculamos la probabilidad de que un estudiante pese menos de 75 kg utilizando la función de distribución acumulativa (CDF)
prob_menos_75kg = norm.cdf(peso_limite_75, media, desv_std)

# Cálculo del número de estudiantes que pesan menos de 75 kg
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("Número de estudiantes que pesan menos de 75 kg:", estudiantes_menos_75kg)

# Graficar la distribución normal con el área sombreada
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.figure(figsize=(6, 4))
plt.plot(x, y, 'b', linewidth=2)
# Rellenar el área bajo la curva desde el inicio hasta el peso_limite
plt.fill_between(x, y, where=(x <= peso_limite_75), color='skyblue', alpha=0.5)
plt.title(f'Probabilidad de que un estudiante pese menos de 75 kg. ($\mu={media}$, $\sigma={desv_std}$, $N={total_estudiantes}$)')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(6, 4))

# Parámetros de la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
total_estudiantes = 500

# VARIABLE para Cálculo de la probabilidad de que un estudiante pese menos de 75 kg
peso_limite = 62

# Primero necesitamos calcular la probabilidad de que un estudiante pese menos
# de 75 kg utilizando la función de distribución acumulativa (CDF) de la distribución normal
prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)

# Cálculo del número de estudiantes que pesan menos de 75 kg
estudiantes_menos_62kg = total_estudiantes * prob_menos_62kg

print("Probabilidad de que un estudiante pese menos de 62 kg:", prob_menos_62kg)
print("Número de estudiantes que pesan menos de 75 kg:", estudiantes_menos_62kg)

# Calcular la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Graficar la Distribución normal
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 62 kg, \n $\mu=70$, $\sigma=4$, $N=500$')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))

# Parámetros de la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
peso_limite = 79 # Peso límite
total_estudiantes = 500

# Calculamos la probabilidad de que un estudiante pese menos o igual a 79 kg
prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)

# La probabilidad de que un estudiante pese más de 79 kg es complementaria
prob_mas_79kg = 1 - prob_menos_79kg

# Calculamos el número aproximado de estudiantes que pesan más de 79 kg
estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg

print("Probabilidad de que un estudiante pese más de 79 kg:", prob_mas_79kg)
print("Número aproximado de estudiantes que pesan más de 79 kg:", estudiantes_mas_79kg)

# Calcular la densidad de probabilidad de la Distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

# Graficar la Distribución normal
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese más de 79 kg. \n $\mu=70$, $\sigma=4$, $N=500$')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
