```python
from sklearn.linear_model import LinearRegression # Modelo de regresión lineal
import numpy as np

# Datos de entrada
periodos = [2018, 2019, 2020, 2021, 2022, 2023] # No se usa en la regresión, solo como contexto
inversion = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1) # Variable independiente (X)
ganancias = [20, 25, 34, 30, 40, 31] # Variable dependiente (Y)

# Crear y entrenar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(inversion, ganancias)

# Predecir la ganancia para una inversión de 8 millones
ganancia_2024 = modelo.predict([[8]])
print("La ganancia esperada para el año 2024 es de:", ganancia_2024[0])

# Coeficientes de la regresión
print("Coeficiente de la pendiente (m):", modelo.coef_[0])
print("Intercepcion de la recta (b):", modelo.intercept_)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Datos de entrada
publicidad = np.array([2, 3, 5, 4, 11, 5]).reshape(-1, 1) # Variable independiente (X)
ganancias = np.array([20, 25, 34, 30, 40, 31]) # Variable dependiente (Y)

# Crear y entrenar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(publicidad, ganancias)

# Predecir la ganancia para una inversión de 0 en publicidad
inversion_prediccion = np.array([[0]])
ganancia_predicha = modelo.predict(inversion_prediccion)

print("La ganancia esperada para una inversión de 0 en publicidad es:", ganancia_predicha[0])
print("Valor de la intercepción:", modelo.intercept_)
print("Valor de la pendiente:", modelo.coef_[0])

# Crear el diagrama de dispersión con la línea de regresión
plt.figure(figsize=(6, 5))
plt.scatter(publicidad, ganancias, color='blue', label='Datos')

# Para graficar la línea de regresión correctamente:
# 1. Se predicen los valores Y para los valores X originales
ganancia_predicha_linea = modelo.predict(publicidad)
# 2. Se usa esta nueva serie de predicciones para el plot.
plt.plot(publicidad, ganancia_predicha_linea, color='red', label='Línea de regresión')
# Nota: La línea de código en la diapositiva original (Slide 20) usaba la variable 'ganancia_predicha' (que solo tiene la predicción para X=0), lo cual causaría un error. He corregido la variable de la línea para asegurar que el gráfico se dibuje correctamente, siguiendo la lógica de la imagen.

# Punto de la predicción (0, y_predicha)
plt.scatter(inversion_prediccion, ganancia_predicha, color='green', marker='x', s=100,
            label='Predicción (0, {0:.2f})'.format(ganancia_predicha[0]))

plt.title('Pronóstico de Ganancias a partir de la Inversión en Publicidad')
plt.xlabel('Inversión en Publicidad (millones)')
plt.ylabel('Ganancias Anuales (millones)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import pearsonr

# Datos de entrada
publicidad = [2, 3, 5, 4, 11, 5]
ganancias = [20, 25, 34, 30, 40, 31]

# Calcular el coeficiente de correlación de Pearson
coeficiente_correlacion, p_valor = pearsonr(publicidad, ganancias)

print("Coeficiente de correlación de Pearson:", coeficiente_correlacion)
print("Valor p:", p_valor)

# Resultado impreso en la diapositiva:
# Coeficiente de correlación de Pearson: 0.9090909090909092
# Valor p: 0.012021076814425222
```
